## TEMPLATE MATCHING
Find objects in an image using Template Matching

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# larger image
img_large = cv2.imread('rubik.png')

# small image or template image
img_small = cv2.imread('rubik_crop.png')
```

#### Preview the Image

```python
plt.subplot(121)
plt.imshow(cv2.cvtColor(img_large, cv2.COLOR_BGR2RGB))
plt.title('large image')

plt.subplot(122)
plt.imshow(cv2.cvtColor(img_small, cv2.COLOR_BGR2RGB))
plt.title('template image')
```

```python
print(img_large.shape)
print(img_small.shape)
```

#### Apply the template matching

```python
large_copy = img_large.copy()
res = cv2.matchTemplate(image=large_copy,
                        templ=img_small,
                        method=cv2.TM_CCOEFF)
```

```python
plt.subplot(121)
plt.imshow(res, cmap='hot')

plt.subplot(122)
plt.imshow(cv2.cvtColor(img_large, cv2.COLOR_BGR2RGB))
```

```python
# find the location
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
```

```python
# draw a rectangle
w, h, c = img_small.shape

top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)

cv2.rectangle(img=large_copy, 
              pt1=top_left, 
              pt2=bottom_right, 
              color=(0,0,255), 
              thickness=5);

plt.imshow(cv2.cvtColor(large_copy, cv2.COLOR_BGR2RGB))
```

```python
cv2.imwrite('matching_image.png', large_copy)
```
