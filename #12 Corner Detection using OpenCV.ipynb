```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('sample_image.PNG')
```

```python
plt.figure(figsize=(10,5))
plt.imshow(img, cmap='gray')
```

```python
img.shape
```

### Haris Corner Detection

```python
# convert the data type
img_float = np.float32(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))
```

```python
# corner detection
dst = cv2.cornerHarris(src=img_float, blockSize=2, ksize=3, k=0.04)
```

```python
# preview the output
plt.figure(figsize=(10,5))
plt.imshow(dst,cmap='gray')
```

```python
print(dst.min())
print(dst.max())
```

```python
plt.figure(figsize=(10,5))
plt.imshow(np.clip(dst,0,1),cmap='gray')
```

### Get the Corner Location

```python
# clip the pixel value
dst = np.clip(dst, 0, 1)
```

```python
# corner location
centers = np.where(dst == dst.max())
```

```python
np.shape(centers)
```

```python
centers
```

```python
# draw a circle of single corner location
cv2.circle(img, center=(centers[1][0],centers[0][0]), radius=5, color=(255,0,0), thickness=-1);
```

```python
plt.figure(figsize=(10,5))
plt.imshow(img)
```

```python
# draw circle of all the detected corner
for i in range(len(centers[0][:])):
    cv2.circle(img, center=(centers[1][i],centers[0][i]), radius=5, color=(255,0,0), thickness=-1);
```

```python
plt.figure(figsize=(10,5))
plt.imshow(img)
```
